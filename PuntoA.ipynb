```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from mpl_toolkits.mplot3d import Axes3D
from numpy.fft import fft2, fftshift, ifft2 
import pandas as pd

#genera m치scara circular opaca
def icmask(a,b,radius,N):  # genera complemento de Circ
  t, s = np.ogrid[0:N,0:N]
  mask = (t-N/2)**2 + (s-N/2)**2 <= radius*radius
  array = np.ones((N,N))
  array[mask] = 0.
  return array

#funci칩n para convertir el archivo csv en matriz
def csv_array(csv):
    array = np.zeros((N,N))*np.exp(-1j*0)
    for i in range(0,(N-1)):
        for u in range(0,(N-1)):
            array[i,u]=csv.at[i,u]
    return array



```

```python
#Leer el archivo
archivo=pd.read_csv('a.csv', header=None)

N=len(archivo)


U=csv_array(archivo)


#radios muy grandes limitan las frecuencias espaciales que pasan por el sistema
radio = 10

#se define la funci칩n de transferencia H
H= icmask(int((N)/2),int((N)/2),radio,N)
Hi=H*np.exp(-1j*0)

#se aplica transformada de fourier a la imagen, correspondiente a la primera lente
T1=np.fft.fft2(U)
Tf=np.fft.fftshift(T1)


TFH=Tf*Hi


#se aplica transformada inversa de fourier(en el sistema 칩ptico corresponde a una transformada de fourier- la imagen se ve reflejada)      
Uf2=np.fft.ifft2(TFH)
  
Uf2c=np.fft.fftshift(Uf2)
UU=abs(Uf2*Uf2)
UM=abs(U*U)

```
